```python
import pandas as pd
import numpy as np
```

```python
df_tratada = pd.read_excel("oficial.xlsx", sheet_name="Tratada")
```

```python
df_tratada['BP_Systolic'] = df_tratada['Blood Pressure'].str.split('/').str[0].astype(float)
df_tratada['BP_Diastolic'] = df_tratada['Blood Pressure'].str.split('/').str[1].astype(float)
```

```python
colunas = [
    'Age', 'Cholesterol', 'Heart Rate', 'BP_Systolic', 'BP_Diastolic',
    'Triglycerides', 'Exercise Hours Per Week', 'Sedentary Hours Per Day',
    'BMI', 'Income', 'Stress Level', 'Physical Activity Days Per Week',
    'Sleep Hours Per Day'
]

resultados = []
```

```python
for col in colunas:
    Q1 = df_tratada[col].quantile(0.25)
    Q3 = df_tratada[col].quantile(0.75)
    IQR = Q3 - Q1
    limite_inf = Q1 - 1.5 * IQR
    limite_sup = Q3 + 1.5 * IQR
    
    outliers = df_tratada[(df_tratada[col] < limite_inf) | (df_tratada[col] > limite_sup)]
    n_outliers = len(outliers)
    
    resultados.append({
        'Coluna': col,
        'Q1': round(Q1, 2),
        'Q3': round(Q3, 2),
        'IQR': round(IQR, 2),
        'Limite Inferior': round(limite_inf, 2),
        'Limite Superior': round(limite_sup, 2),
        'Outliers': n_outliers
    })
    
    # Remove outliers da base
    df_tratada = df_tratada[(df_tratada[col] >= limite_inf) | (df_tratada[col].isna())]
    df_tratada = df_tratada[(df_tratada[col] <= limite_sup) | (df_tratada[col].isna())]
```

```python
# Exibe resultado
resultado_df = pd.DataFrame(resultados)
print(resultado_df.to_string(index=False))
print(f"\nRegistros antes: 2652")
print(f"Registros após remoção: {len(df_tratada)}")
print(f"Removidos: {2652 - len(df_tratada)}")
```
