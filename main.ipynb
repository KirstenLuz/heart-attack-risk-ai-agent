```python
def faixa_renda(valor):
    if valor < 50000:
        return 1   # baixa
    elif valor < 100000:
        return 2   # média-baixa
    elif valor < 150000:
        return 3   # média
    elif valor < 200000:
        return 4   # média-alta
    elif valor < 250000:
        return 5   # alta
    else:
        return 6   # muito alta
    

my_df["Renda"] = df_original["Renda"].apply(faixa_renda)
```

```python
print(my_df["Renda"])
```

```python
def faixa_imc(valor):
    if valor < 18.5:
        return 1
    elif valor < 25:
        return 2
    elif valor < 30:
        return 3
    elif valor < 35:
        return 4
    elif valor < 40:
        return 5
    else:
        return 6

my_df["IMC"] = df_original["IMC"].apply(faixa_imc)
```
